```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
data.shape
```

```python
data_dup = data.duplicated().any()
data_dup
```

```python
data = data.drop_duplicates()
data.shape
```

```python
category_val = []
content_val = []

for column in data.columns:
    if data[column].nunique() <=10:
        category_val.append(column)
    else:
        content_val.append(column)
```

```python
category_val
```

```python
content_val
```

```python
category_val.remove('sex')
category_val.remove('target')
data = pd.get_dummies(data, columns=category_val, drop_first=True)
```

```python
data.head()
```

```python
st = StandardScaler()
data[content_val] = st.fit_transform(data[content_val])
```

```python
data.head()
```

```python
Xi = data.drop(['target'], axis=1)
yi = data['target']

X_tr, X_te, y_tr, y_te = train_test_split(Xi, yi, test_size=0.18, random_state=30)
```

```python
lor = LogisticRegression()
lor.fit(X_tr, y_tr)

svm = svm.SVC()
svm.fit(X_tr, y_tr)

knn = KNeighborsClassifier()
knn.fit(X_tr, y_tr)

dt = DecisionTreeClassifier()
dt.fit(X_tr, y_tr)
```

```python
y_pred1 = lor.predict(X_te)
y_pred2 = svm.predict(X_te)
y_pred3 = knn.predict(X_te)
y_pred4 = dt.predict(X_te)
```

```python
score1 = accuracy_score(y_te, y_pred1)
score2 = accuracy_score(y_te, y_pred2)
score3 = accuracy_score(y_te, y_pred3)
score4 = accuracy_score(y_te, y_pred4)
```

```python
print(score1, score2, score3, score4)
```
